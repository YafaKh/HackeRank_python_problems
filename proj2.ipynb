```python
import pandas as pd
df = pd.read_csv('owid-covid-data.csv', low_memory=False)

num_rows = df.shape[0]  # Number of rows
num_cols = df.shape[1]  # Number of columns

print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
print("head: \n", df.head(10))
print("tail: \n", df.tail())
```

```python
df.describe()
```

```python
df = df.drop(columns= ['new_deaths_smoothed', 'new_cases_per_million', 'total_cases_per_million'])
```

```python
df = df.rename(columns={'date': 'Date', 'location': 'Country', 'continent': 'Continent', 'iso_code': 'ISO_code'})
```

```python
print(df['Continent'].unique().tolist())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='constant')
df2 = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
df.head()
df2.head()
```

```python
df3 = df2.groupby(['Date','Country',])[['Date','Country','total_cases','total_deaths','total_vaccinations']].sum().reset_index()
# it results an error! ValueError: cannot insert Country, already exists
```

```python
columns_to_replace = ['total_cases', 'total_deaths', 'total_vaccinations']
df2[columns_to_replace] = df2[columns_to_replace].replace('missing_value', 0)

df3 = df2.groupby(['Date', 'Country'])[['total_cases', 'total_deaths', 'total_vaccinations']].sum().reset_index()
df3.tail(10)
```
